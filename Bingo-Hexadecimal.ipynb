```
# FUNÇÃO: Conversão Decimal-Hexadecimal ()
def dec_para_hex(num_Decimal):
  return hex(num_Decimal).split('x')[-1]
```

```
# FUNÇÃO: Sortear uma (1) Pedra: RANDOM DECIMAL: [16, 159]
def sortear_pedra():
   while True:
    sortear = np.random.randint(16, 159, 1)
    if sortear in bd_pedras:
      sortear = np.random.randint(16, 159, 1)
    else:
      bd_pedras[np.count_nonzero(bd_pedras)] = sortear
      sortear1 = dec_para_hex(sortear[0])
      return sortear1
      break
```

```
# FUNÇÃO: Exibir Pontuação
def pontuacao(_cartela_1, _cartela_2, _cartela_3, nominal = False):
  if nominal == False:
    _tabela1 = _cartela_1
    total_x1 = np.count_nonzero(_tabela1 == 'x')

    _tabela2 = _cartela_2
    total_x2 = np.count_nonzero(_tabela2 == 'x')

    _tabela3 = _cartela_3
    total_x3 = np.count_nonzero(_tabela3 == 'x')
    
    jogador_1 = (total_x1 / 16) * 100
    jogador_2 = (total_x2 / 16) * 100
    jogador_3 = (total_x3 / 16) * 100
    
    return np.array([[jogador_1,'jogador1'],[jogador_2,'jogador2'],[jogador_3,'jogador3']])
  else:
    _tabela1 = _cartela_1[0]
    total_x1 = np.count_nonzero(_tabela1 == 'x')

    _tabela2 = _cartela_2[0]
    total_x2 = np.count_nonzero(_tabela2 == 'x')

    _tabela3 = _cartela_3[0]
    total_x3 = np.count_nonzero(_tabela3 == 'x')
    
    jogador_1 = (total_x1 / 16) * 100
    jogador_2 = (total_x2 / 16) * 100
    jogador_3 = (total_x3 / 16) * 100
    
    return np.array([[jogador_1,_cartela_1[1]],[jogador_2,_cartela_2[1]],[jogador_3,_cartela_3[1]]])
```

## ATENÇÃO - PROBLEMA :-|  : O método RANDINT repeti números já sorteados.

&#10139;  **IMPORTANTE**: Temos que resolver este problema ? Criar Função Nova !!!

```
def sortearSemRepeticao(INICIO, FIM, TAMANHO):
  # Criar uma função que crie um conjunto de dados (dataset), sem valores repetidos !!!
  # Inserir no dataset, somente números não repetidos dentro do intervalo [INICIO, FIM]
  numSorteados = np.zeros(TAMANHO, dtype=int)
  contagem = 0
  i = -1
  while (contagem < TAMANHO):
    for sorteado in np.random.randint(INICIO, FIM, 1):
      if (sorteado not in numSorteados):
        contagem += 1
        i += 1
        numSorteados[i] = sorteado
  return numSorteados
```
